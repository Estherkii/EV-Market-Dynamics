### separating the year, brand, and model from model column

```python
def extract_details_and_add(df, column_name):    
    pattern = r'(?P<Year>\d{4}) (?P<Brand>\w+) (?P<Model>.*?)( \d+|$)'

    details = df[column_name].str.extract(pattern, expand=True)

    df['Year'] = details['Year']
    df['Brand'] = details['Brand']
    df['Model'] = details['Model']

    return df
```

```python
ev_prices = extract_details_and_add(ev_prices, 'Model')
ev_prices
```

### removing 'Tax Credit' column

```python
ev_prices = ev_prices.drop('Tax Credit', axis=1)
```

```python
ev_prices.head()
```

### changing column names

```python
ev_prices = ev_prices.rename(columns={
    'Model': 'model',
    'Base Price': 'base_price',
    'Dest. Charge': 'delivery_charge',
    'Effective Price': 'final_price',
    'Year': 'year',
    'Brand': 'brand'
})
```

### cleaning prices columns

```python
columns_to_clean ={
    'base_price', 'delivery_charge', 'final_price'
}
```

```python
def clean_price_columns(df, columns):
    for column in columns:
        
        df[column] = df[column].astype(str).str.replace(r'[^0-9.-]', '', regex=True)

        df[column] = pd.to_numeric(df[column], errors='coerce')

    return df


ev_prices = clean_price_columns(ev_prices, columns_to_clean)
```

# Cleaning Ev Sales and Stock per Year

```python
ev_sales_scraping = pd.read_csv(r"C:\Users\photo\Desktop\Ironhack\projects\EV-Market-Dynamics\data\scraping\wikipedia_ev_sales_scraping.csv", encoding='latin1')
ev_sales_scraping.head()
```

#### Modifying columns

```python
ev_sales_scraping.columns = [col.lower().replace(' ', '_') for col in ev_sales_scraping.columns]
```

```python
def remove_brackets(text):
    return re.sub(r"\[.*?\]", "", text) if isinstance(text, str) else text

for column in ev_sales_scraping.columns:
    ev_sales_scraping['country'] = ev_sales_scraping['country'].apply(remove_brackets)
```

```python
ev_sales_scraping.head()
```

## diving the df in two tables, one with stock and other one with sales
###### first cleaning row Europe

```python
ev_sales_scraping.loc[ev_sales_scraping['country'] == 'Europe (EU+3)', 'country'] = 'Europe'
```

```python
valid_countries = [
    'United States', 'Germany', 'France', 'United Kingdom',
    'Norway', 'Netherlands', 'Sweden', 'Japan', 'Canada', 'China', 'Europe'
]

ev_sales_scraping = ev_sales_scraping[ev_sales_scraping['country'].isin(valid_countries)]

ev_sales_scraping.reset_index(drop=True, inplace=True)
```

```python
ev_sales_scraping.head()
```

###### dividing stock

```python
columns_to_include = [col for col in ev_sales_scraping.columns if '.1' not in col]

ev_stock = ev_sales_scraping[columns_to_include]
ev_stock.head()
```

```python
def clean_numeric(value):
  
    cleaned_value = re.sub(r'\[.*?\]', '', str(value))
    cleaned_value = re.sub(r'[^\d.]+', '', cleaned_value)
    

    if not cleaned_value or not any(char.isdigit() for char in cleaned_value):
        return None

    try:
        
        return int(cleaned_value)
    except ValueError:
        return float(cleaned_value)
```

```python
for col in ev_stock.columns:
    if col != 'country':
        ev_stock.loc[:, col] = ev_stock[col].apply(clean_numeric)
```

```python
ev_stock
```

```python
ev_stock.dtypes
```

```python
ev_stock = ev_stock.copy()
```

```python
for col in ev_stock.columns:
    if col != 'country':
        ev_stock[col] = ev_stock[col].astype(float)

print(ev_stock.dtypes)
```
