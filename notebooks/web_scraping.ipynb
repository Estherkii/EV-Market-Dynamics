# Getting EV prices information from website

```python
url = "https://insideevs.com/news/565883/electric-car-prices-us/"
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")
```

```python
tbody = soup.find('tbody')
```

```python
rows = tbody.find_all('tr')
```

```python
all_rows = []

for row in tbody.find_all('tr'):
    cols = row.find_all('td')
    cols_text = [ele.text.strip() for ele in cols]
    all_rows.append(cols_text)
```

```python
headers = all_rows[0]
data_rows = all_rows[1:]
```

```python
ev_prices_scraping = pd.DataFrame(data_rows, columns=headers)
ev_prices_scraping.head()
```

# Scraping updated prices

```python
url = "https://ev-database.org/#sort:path~type~order=.rank~number~desc|range-slider-range:prev~next=0~1200|range-slider-acceleration:prev~next=2~23|range-slider-topspeed:prev~next=110~350|range-slider-battery:prev~next=10~200|range-slider-towweight:prev~next=0~2500|range-slider-fastcharge:prev~next=0~1500|paging:currentPage=0|paging:number=all"
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")
```

```python
ev_prices2 = []

ev_elements = soup.find_all('div', class_='list-item')
```

```python
def extract_car_data(ev_elements):
    cars_list = []

    for wrapper in ev_elements:
        
        car_brand_element = wrapper.find('span', class_=True)
        car_brand = car_brand_element.get('class')[0] if car_brand_element else 'Brand not found'
        
        car_model_element = wrapper.find('span', class_='model')
        car_model = car_model_element.get_text(strip=True) if car_model_element else 'Model not found'
        
        price_de_element = wrapper.find('span', class_='country_de')
        price_de = price_de_element.get_text(strip=True) if price_de_element else 'DE Price not found'

        price_uk_element = wrapper.find('span', class_='country_uk')
        price_uk = price_uk_element.get_text(strip=True) if price_uk_element else 'UK Price not found'

        price_nl_element = wrapper.find('span', class_='country_nl')  
        price_nl = price_nl_element.get_text(strip=True) if price_nl_element else 'NL Price not found'

        car_acceleration_element = wrapper.find('span', class_='acceleration')  
        car_acceleration = car_acceleration_element.get_text(strip=True) if car_acceleration_element else 'Acceleration not found'

        car_top_speed_element = wrapper.find('span', class_='topspeed')  
        car_top_speed = car_top_speed_element.get_text(strip=True) if car_top_speed_element else 'Top speed not found'

        car_range_element = wrapper.find('span', class_='erange_real')  
        car_range = car_range_element.get_text(strip=True) if car_range_element else 'Range not found'

        car_efficiency_element = wrapper.find('span', class_='efficiency')  
        car_efficiency = car_efficiency_element.get_text(strip=True) if car_efficiency_element else 'Efficiency not found'

        car_fastcharge_element = wrapper.find('span', class_='fastcharge_speed_print')  
        car_fastcharge = car_fastcharge_element.get_text(strip=True) if car_fastcharge_element else 'Fastcharge not found'
        
        car_towing_capacity_element = wrapper.find('span', class_='towweight')  
        car_towing_capacity = car_towing_capacity_element.get_text(strip=True) if car_towing_capacity_element else 'Towing capacity not found' 
        
        car_battery_element = wrapper.find('span', class_='battery')
        car_battery = car_battery_element.get_text(strip=True) if car_battery_element else 'Useability Battery not found'
        
        car_v2_load_element = wrapper.find('span', title='Vehicle-2-Load Bi-directional charging possible')
        car_v2_load = car_v2_load_element.get_text(strip=True) if car_v2_load_element else 'No'
        
        car_v2_home_element = wrapper.find('span', title = 'Vehicle-2-Home Bi-directional charging possible')
        car_v2_home = car_v2_home_element.get_text(strip=True) if car_v2_home_element else 'No'
        
        car_v2_grid_element = wrapper.find('span', title = 'Vehicle-2-Grid Bi-directional charging possible')
        car_v2_grid = car_v2_grid_element.get_text(strip=True) if car_v2_grid_element else 'No'
            
        car_availability_element = wrapper.find('span', class_='current')  
        car_availability = car_availability_element.get_text(strip=True) if car_availability_element else 'Availability not found' 

        car_data = {
            'brand': car_brand,
            'model': car_model,
            'de_price': price_de,
            'uk_price': price_uk,
            'nl_price': price_nl,
            'acceleration_0-100km/h': car_acceleration,
            'top_speed': car_top_speed,
            'range': car_range,
            'efficiency': car_efficiency,
            'fastcharge': car_fastcharge,
            'towing_capacity(kg)': car_towing_capacity,
            'useable_battery': car_battery,
            'v2l' : car_v2_load,
            'v2h' : car_v2_home,
            'v2g' : car_v2_grid,
            'availability': car_availability
        }
        cars_list.append(car_data)

    return cars_list

def print_compact_car_data(cars_list):
    for car in cars_list:
        print(f"Name: {car['name']}, DE Price: {car['de_price']}, Range: {car['range']}")
```
