### separating the year, brand, and model from model column

```python
def extract_details_and_add(df, column_name):    
    pattern = r'(?P<Year>\d{4}) (?P<Brand>\w+) (?P<Model>.*?)( \d+|$)'

    details = df[column_name].str.extract(pattern, expand=True)

    df['Year'] = details['Year']
    df['Brand'] = details['Brand']
    df['Model'] = details['Model']

    return df
```

```python
ev_prices = extract_details_and_add(ev_prices, 'Model')
ev_prices
```

### removing 'Tax Credit' column

```python
ev_prices = ev_prices.drop('Tax Credit', axis=1)
```

```python
ev_prices.head()
```

### changing column names

```python
ev_prices = ev_prices.rename(columns={
    'Model': 'model',
    'Base Price': 'base_price',
    'Dest. Charge': 'delivery_charge',
    'Effective Price': 'final_price',
    'Year': 'year',
    'Brand': 'brand'
})
```

### cleaning prices columns

```python
columns_to_clean ={
    'base_price', 'delivery_charge', 'final_price'
}
```

```python
def clean_price_columns(df, columns):
    for column in columns:
        
        df[column] = df[column].astype(str).str.replace(r'[^0-9.-]', '', regex=True)

        df[column] = pd.to_numeric(df[column], errors='coerce')

    return df


ev_prices = clean_price_columns(ev_prices, columns_to_clean)
```

```python
def clean_numeric(value):
  
    cleaned_value = re.sub(r'\[.*?\]', '', str(value))
    cleaned_value = re.sub(r'[^\d.]+', '', cleaned_value)
    

    if not cleaned_value or not any(char.isdigit() for char in cleaned_value):
        return None

    try:
        
        return int(cleaned_value)
    except ValueError:
        return float(cleaned_value)
```

```python
for col in ev_stock.columns:
    if col != 'country':
        ev_stock.loc[:, col] = ev_stock[col].apply(clean_numeric)
```

###### cleaning de_price, uk_price, nl_price, acceleration, top_speed, range, efficiency and fastcharge. Also changing to int

```python
def clean_and_convert_to_int_with_nan(df, columns):
  
    for column in columns:
      
        df[column] = df[column].str.replace('[^\d]', '', regex=True)

        df[column] = pd.to_numeric(df[column], errors='coerce').astype('Int64')

    return df


columns_to_clean = ['de_price', 'uk_price', 'nl_price', 'acceleration_0-100km/h', 'top_speed', 'range', 'efficiency', 'fastcharge' ]
ev_prices23 = clean_and_convert_to_int_with_nan(ev_prices23, columns_to_clean)
```

```python
ev_prices23.dtypes
```

#### Checking and handeling nulls

```python
ev_prices23.isnull().sum()
```

##### Eliminate rows where only there are null values and we do not have brand and model

```python
ev_prices23 = ev_prices23.dropna(subset=['brand'])
```

```python
ev_prices23.isnull().sum()
```

###### changing towing_capacity from float to int

```python
ev_prices23['towing_capacity(kg)'].fillna(0, inplace=True) 

ev_prices23['towing_capacity(kg)'] = ev_prices23['towing_capacity(kg)'].astype(int)
```

###### Removing extra information in model cars

```python
#replacing weird characters and (EDITION for nothing
ev_prices23['model'] = ev_prices23['model'].replace(r"\(.*?\)|\(EDITION", "", regex=True)

#replacing commas
ev_prices23['model'] = ev_prices23['model'].str.replace(',', '', regex=False)

# replacing word AT for nothing
ev_prices23['model'] = ev_prices23['model'].str.replace('AT', '', case=False, regex=False)

#deletinb blank spaces at the end of the model
ev_prices23['model'] = ev_prices23['model'].str.rstrip()
```

```python
ev_prices23['year'] = '2023'
```

```python
ev_prices23.rename(columns={'range ': 'car_range', 'acceleration_0-100km/h': 'acceleration', 'towing_capacity(kg)': 'towing_capacity' }, inplace=True)
```
